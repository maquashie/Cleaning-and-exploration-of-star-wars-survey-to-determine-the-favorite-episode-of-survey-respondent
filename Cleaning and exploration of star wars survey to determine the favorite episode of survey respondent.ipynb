# Code to read dataset (Star_wars)
encoding specified because the data set has some characters that aren't in Python's default utf-8 encoding

```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

# Explored data to look for any strange values

```python
 star_wars.head(10)
```

# Found a couple of NAN for respondent ID so will be removed

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
star_wars.head(10)
```

# Replaced yes and No in the second and third columns with boolean values

```python
yes_no = {
    "Yes": True,
    "No": False
}
selected_cols= [
    "Have you seen any of the 6 films in the Star Wars franchise?",
    "Do you consider yourself to be a fan of the Star Wars film franchise?"
    ]
for col in selected_cols:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
```

# Converted the next six collumns into boolean and renamed them so the names are more intuitive

```python
import  numpy as np


movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)

    
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })

star_wars.head()
```

# Converting column 10 to 15 to numeric and rename them to reflect what they represent.

```python
star_wars = star_wars.rename(columns={
        "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
        "Unnamed: 10": "ranking_2",
        "Unnamed: 11": "ranking_3",
        "Unnamed: 12": "ranking_4",
        "Unnamed: 13": "ranking_5",
        "Unnamed: 14": "ranking_6"
        })
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars.head()
```

# Computing of rankings

```python
star_wars[star_wars.columns[9:15]].mean()
```

# Bar plots for rankings

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.bar(range(6), star_wars[star_wars.columns[9:15]].mean())
```

```python
star_wars[star_wars.columns[3:9]].sum()
```

```python
plt.bar(range(6), star_wars[star_wars.columns[3:9]].sum())
```

More respondents appeared to have seen original movies. This corresponds to the plot for the rankings.

# Analysing movies by gender

The data is split into male and female

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python

plt.bar(range(6),  males[males.columns[9:15]].mean())
plt.title("male respondents")
plt.show()

plt.bar(range(6),  females[females.columns[9:15]].mean())
plt.title("female respondents")
plt.show()
```

```python
plt.bar(range(6),  males[males.columns[9:15]].sum())
plt.title("male respondents")
plt.show()

plt.bar(range(6),  females[females.columns[9:15]].sum())
plt.title("female respondents")
plt.show()
```

# Observation
From the plots, it could be observed that,there were  more male respondents than female respondents. Also, the most viewed and liked movie for both genders appear to be episode III.

# Exploring the education column and segmenting them based on respondents diploma (high school, associated degree, bachelors, etc.)

```python
star_wars["Education"].head(15)
```

```python
star_wars = star_wars[pd.notnull(star_wars["Education"])]
star_wars["Education"].head(15)
```

```python
Respondents_with_HS = star_wars[star_wars["Education"] == "High school degree"]
Respondents_with_AD = star_wars[star_wars["Education"] == "Some college or Associate degree"]
Respondents_with_BD = star_wars[star_wars["Education"] == "Bachelor degree"]
Respondents_with_GD = star_wars[star_wars["Education"] == "Graduate degree"]
```

```python
plt.bar(range(6),  Respondents_with_HS[Respondents_with_HS.columns[9:15]].mean())
plt.title("Respondents_with_HS")
plt.show()

plt.bar(range(6),  Respondents_with_AD[Respondents_with_AD.columns[9:15]].mean())
plt.title("Respondents_with_AD")
plt.show()

plt.bar(range(6),  Respondents_with_BD[Respondents_with_BD.columns[9:15]].mean())
plt.title("Respondents_with_BD")
plt.show()

plt.bar(range(6),  Respondents_with_GD[Respondents_with_GD.columns[9:15]].mean())
plt.title("Respondents_with_GD")
plt.show()
```

It can be observed from the above plots that movie preference of respondents differ based on their education level. Whiles most respondents with high school degree prefered episode VI, those with graduate degree prefered episode III of star wars.
